```python
import urllib3
import itertools
import numpy as np
```

```python
# import word list

http = urllib3.PoolManager()
target_url = "http://www.sil.org/linguistics/wordlists/english/wordlist/wordsEn.txt"
r = http.request('GET',target_url)
wordlist = str(r.data.decode('utf-8')).split('\r\n')
```

```python
### Here we make a function that tells us whether or not its an english word

english_words = set(word.strip().lower() for word in wordlist)
def is_english_word(word):
    if word.lower() in english_words:
        print('"'+word + '" is a word')
        return 
        
   
    else:
        return

is_english_word("ham")
```

```python
print(is_english_word('larvaase'))
```

```python
is_english_word('larvaase')
```

```python
def string_combos(string):
    num = len(string)
    ans = list()
    for t in range(0,num):
        num2 = num-t
        itr1 = itertools.permutations(string,num2)
        for itr in itr1:
            wrd = ''.join(itr)
            ans.append(wrd)
    return ans
```

```python
string_combos('abcd')
```
